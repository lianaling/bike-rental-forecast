# 1.0 Data Preprocessing

In this section, the raw dataset will be load and cleaned. The raw dataset load are the hour.csv and day.csv.

## 1.1 Import libary

```python
import pandas as pd
import numpy as np
```

## 1.2 Load data

```python
df_hour = pd.read_csv("../dataset/hour.csv")
df_day = pd.read_csv("../dataset/day.csv")
```

```python
df_hour
```

```python
df_day
```

## 1.3 Combining dataset

```python
df_concatenated = pd.concat([df_hour, df_day])
```

```python
df_concatenated
```

## 1.4 Renaming column

```python
df_concatenated = df_concatenated.rename(
                        columns={'instant':'index',
                                 'dteday':'date',
                                 'yr':'year',
                                 'mnth':'month',
                                 'hr':'hour',
                                 'weathersit':'weather',
                                 'temp':'temperature',
                                 'atemp':'temperaturefelt',
                                 'hum':'humidity',
                                 'casual':'casualuser',
                                 'registered':'registereduser',
                                 'cnt':'count'})
df_concatenated.head()
```

## 1.5 Droping column

```python
df_cleaned = df_concatenated.drop(columns=['index'])
df_cleaned
```

## 1.6 Strategise datatype

```python
df_cleaned.dtypes
```

```python
cols = ['season', 'year','month','hour','holiday','weekday','workingday','weather']

for col in cols:
    df_cleaned[col] = df_cleaned[col].astype('string')
df_cleaned.info()
```

## 1.7 Replacing unrecognised

```python
# seasons = ['winter','spring','summer','fall']
# years = ['2011','2012']
# months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
# weekday = ['']
# workingday = ['']
# weather = ['']

# To replace unrecognised data in column with recognised value
df_cleaned["season"].replace({"1": "winter", "2": "spring", "3": "summer", "4": "fall"}, inplace=True)
df_cleaned["year"].replace({"0": "2011", "1": "2012"}, inplace=True)
df_cleaned["month"].replace({"1": "JAN", "2": "FEB", "3": "MAR", "4": "APR", 
                             "5": "MAY", "6": "JUN", "7": "JUL", "8": "AUG", 
                             "9": "SEP", "10": "OCT", "11": "NOV", "12": "DEC"}, inplace=True)
df_cleaned["hour"].replace({np.NaN: "NaN"}, inplace=True)
```

```python
df_cleaned
```

## 1.8 Check missing

```python
pd.DataFrame(df_cleaned.isnull().sum(), columns=["Count"])
```

## 1.9 Save the file

```python
# save cleaned test data to csv file for later use
df_day = pd.read_csv("../dataset/day.csv")
export_csv = df_cleaned.to_csv("../dataset/cleaned.csv", index = None, header=True)
df_cleaned
```
