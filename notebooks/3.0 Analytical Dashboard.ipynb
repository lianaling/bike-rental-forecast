This was a analytic dashboard used to demonstrate the result. The cleaned dataset load is the cleaned.csv.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
pd.options.display.max_columns = 999
```

```python
# Initialise dashboard
import ipywidgets as widgets
from IPython.display import display, clear_output

# !jupyter nbextension enable --py widgetsnbextension --sys-prefix
# !jupyter serverextension enable voila --sys-prefix

# load data
# df_cleaned = pd.read_csv("../dataset/cleaned.csv")
```

```python
# Setup header
file = open("../resource/widget_setup/header_1.png", "rb")
image = file.read()

image_headline = widgets.Image(
    value=image,
    format='jpg',
    width='800')

header = widgets.HBox([image_headline], description="content",
                      layout=widgets.Layout(width='auto', height="auto")
                     )
                
```

```python
# Setup footer
footer_headertext = widgets.HTML(value="<h2>Observation: </h2>")
```

```python
# Setup content
a1_description = "Count of bikes based on weekdays and weekends"
a2_description = "Count of bikes based on different weathers"
a3_description = "Count of bikes based on different seasons"
a4_description = "Count of bikes based on different months"


file_a1 = open("../resource/fig/a1.png", "rb")
img_a1 = file_a1.read()
file_a2 = open("../resource/fig/a2.png", "rb")
img_a2 = file_a2.read()
file_a3 = open("../resource/fig/a3.png", "rb")
img_a3 = file_a3.read()
file_a4 = open("../resource/fig/a4.png", "rb")
img_a4 = file_a4.read()
               

# Content A: Various Analysis Approach
a1 = widgets.Button(description=a1_description,
                    tooltip='Approach1',
                    layout=widgets.Layout(width="auto", height="auto"),
                    button_style="success")
a2 = widgets.Button(description=a2_description,
                    tooltip='Approach2',
                    layout=widgets.Layout(width="auto", height="auto"),
                    button_style="primary")
a3 = widgets.Button(description=a3_description,
                    tooltip='Approach3',
                    layout=widgets.Layout(width="auto", height="auto"),
                    button_style="info")
a4 = widgets.Button(description=a4_description,
                    tooltip='Approach4',
                    layout=widgets.Layout(width="auto", height="auto"),
                    button_style="warning")

content_A = widgets.TwoByTwoLayout(top_left=a1,
                                   top_right=a2,
                                   bottom_left=a3,
                                   bottom_right=a4)

# Content B: Selected Approach & Output
output = widgets.Output()
imge_to_show1 = widgets.Image(
    value=img_a1,
    format='png',
    width='auto')
imge_to_show2 = widgets.Image(
    value=img_a2,
    format='png',
    width='auto')
imge_to_show3 = widgets.Image(
    value=img_a3,
    format='png',
    width='auto')
imge_to_show4 = widgets.Image(
    value=img_a4,
    format='png',
    width='auto')

def on_a1_clicked(event):
    with output:
        clear_output()
        try:
            selected_approach = widgets.HTML(value="<h3>Selected analysis approach: "+
                                             "<i style='color:red;'>" + a1_description +
                                             "</i>" + " </h3>")
            display(selected_approach)
            img_to_show = img_a1
            display(imge_to_show1)
            footer_result = "Most rider rent the bike from 18:00 to 19:00. " + "Bike rental during weedays are greater than weenkends.\"\n"
            footer_content = widgets.HTML(value="<p>" + footer_result + "</p>")
            footer = widgets.VBox([footer_headertext, footer_content], description="content",
                                  layout=widgets.Layout(width='auto', height="auto"))
            display(footer)
        except Exception as e:
            print("No analysis available...")
def on_a2_clicked(event):
    with output:
        clear_output()
        try:
            selected_approach = widgets.HTML(value="<h3>Selected analysis approach: "
                                             + "<i style='color:red;'>" + a2_description + "</i>" + " </h3>")
            display(selected_approach)
            img_to_show = img_a2
            display(imge_to_show2)
            footer_result = "Most rider rent the bike from 18:00 to 19:00. " + "Bike rental during weather..."
            footer_content = widgets.HTML(value="<p>" + footer_result + "</p>")
            footer = widgets.VBox([footer_headertext, footer_content], description="content",
                                  layout=widgets.Layout(width='auto', height="auto"))
            display(footer)
        except Exception as e:
            print("No analysis available...")
def on_a3_clicked(event):
    with output:
        clear_output()
        try:
            selected_approach = widgets.HTML(value="<h3>Selected analysis approach: "
                                             + "<i style='color:red;'>" + a3_description + "</i>" + " </h3>")
            display(selected_approach)
            img_to_show = img_a3
            display(imge_to_show3)
            footer_result = "Most rider rent the bike from 18:00 to 19:00. " + "Bike rental during seasons..."
            footer_content = widgets.HTML(value="<p>" + footer_result + "</p>")
            footer = widgets.VBox([footer_headertext, footer_content], description="content",
                                  layout=widgets.Layout(width='auto', height="auto"))
            display(footer)
        except Exception as e:
            print("No analysis available...")
def on_a4_clicked(event):
    with output:
        clear_output()
        try:
            selected_approach = widgets.HTML(value="<h3>Selected analysis approach: "
                                             + "<i style='color:red;'>" + a4_description + "</i>" + " </h3>")
            display(selected_approach)
            img_to_show = img_a4
            display(imge_to_show4)
            footer_result = "Most rider rent the bike from 18:00 to 19:00. " + "Bike rental during months..."
            footer_content = widgets.HTML(value="<p>" + footer_result + "</p>")
            footer = widgets.VBox([footer_headertext, footer_content], description="content",
                                  layout=widgets.Layout(width='auto', height="auto"))
            display(footer)
        except Exception as e:
            print("No analysis available...")

a1.on_click(on_a1_clicked)
a2.on_click(on_a2_clicked)
a3.on_click(on_a3_clicked)
a4.on_click(on_a4_clicked)

content_B = widgets.VBox([output])

# Stacked content
content_headertext = widgets.HTML(value="<h1>Choose an analysis approach~</h1>")
content_reset = widgets.Button(description="Reset",
                               tooltip='Reset',
                               layout=widgets.Layout(width="wrap-content", height="wrap-content"), 
                               button_style="")

def on_reset_clicked(event):
    with output:
        clear_output()
        try:
            print()
        except Exception as e:
            print("No analysis available...")

content_reset.on_click(on_reset_clicked)
content_header = widgets.HBox([content_headertext])
content = widgets.VBox([content_header, content_A, content_B], description="content",
                       layout=widgets.Layout(width='auto', height="auto"))
```

```python
page = widgets.VBox([header, content, content_reset])
display(page)
```
